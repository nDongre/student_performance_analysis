```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("datasets_74977_169835_StudentsPerformance.csv")
```

```python
data.head()
```

```python
#Check for any missing values
data.isnull().sum()
```

```python
#Lets find out the data types in our dataset
```

```python
data.info()
```

```python
#The above information makes it clear that there are 5 categorical variables and 3 numerical variables.
```

```python
data.describe().T
```

```python
#Let us find out distribution of categorical variables
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3)
data["race/ethnicity"].value_counts().plot(kind="pie", autopct='%1.1f%%', figsize = (20,10),ax=axes[0,0])
data["gender"].value_counts().plot(kind="pie", autopct='%1.1f%%', figsize = (20,10),ax=axes[0,1])
data["parental level of education"].value_counts().plot(kind="pie", autopct='%1.1f%%', figsize = (20,10),ax=axes[0,2])
data["lunch"].value_counts().plot(kind="pie", autopct='%1.1f%%', figsize = (20,10),ax=axes[1,0])
data["test preparation course"].value_counts().plot(kind="pie", autopct='%1.1f%%', figsize = (20,10),ax=axes[1,1])
fig.delaxes(ax= axes[1,2]) 
```

```python
#Correlation between numerical variables
```

```python
plt.subplots(figsize=(5,5)) 
p=sns.heatmap(data.corr(), annot=True, linewidth = 0.5)
```

```python
#The above heat map shows us that kids performing well in one subject are usually performing well in other subjects
```

```python
#Now let us find out relationships between individual dependant variable with independant variable
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize = (20,20))
ax1 = sns.violinplot(data["race/ethnicity"],data["math score"],orient = 'v',palette = 'husl',ax = axes[0][0])
ax1 = sns.violinplot(data["gender"],data["math score"],orient = 'v',palette = 'husl',ax = axes[0][1])
ax1 = sns.violinplot(data["parental level of education"],data["math score"],orient = 'v',palette = 'husl',ax = axes[1][0])
ax1 = sns.violinplot(data["lunch"],data["math score"],orient = 'v',palette = 'husl',ax = axes[1][1])
ax1 = sns.violinplot(data["test preparation course"],data["math score"],orient = 'v',palette = 'husl',ax = axes[2][0])
fig.delaxes(ax= axes[2,1]) 
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize = (20,20))
ax1 = sns.violinplot(data["race/ethnicity"],data["reading score"],orient = 'v',palette = 'husl',ax = axes[0][0])
ax1 = sns.violinplot(data["gender"],data["reading score"],orient = 'v',palette = 'husl',ax = axes[0][1])
ax1 = sns.violinplot(data["parental level of education"],data["reading score"],orient = 'v',palette = 'husl',ax = axes[1][0])
ax1 = sns.violinplot(data["lunch"],data["reading score"],orient = 'v',palette = 'husl',ax = axes[1][1])
ax1 = sns.violinplot(data["test preparation course"],data["reading score"],orient = 'v',palette = 'husl',ax = axes[2][0])
fig.delaxes(ax= axes[2,1]) 
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize = (20,20))
ax1 = sns.violinplot(data["race/ethnicity"],data["writing score"],orient = 'v',palette = 'husl',ax = axes[0][0])
ax1 = sns.violinplot(data["gender"],data["writing score"],orient = 'v',palette = 'husl',ax = axes[0][1])
ax1 = sns.violinplot(data["parental level of education"],data["writing score"],orient = 'v',palette = 'husl',ax = axes[1][0])
ax1 = sns.violinplot(data["lunch"],data["writing score"],orient = 'v',palette = 'husl',ax = axes[1][1])
ax1 = sns.violinplot(data["test preparation course"],data["writing score"],orient = 'v',palette = 'husl',ax = axes[2][0])
fig.delaxes(ax= axes[2,1]) 
```

```python
#Distribution of scores against each other.
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(20,20) )
sns.scatterplot(x="reading score", y="math score", data= data, ax = axes[0][0])
sns.scatterplot(x="writing score", y="math score", data= data, ax = axes[0][1])

sns.scatterplot(x="math score", y="reading score", data= data, ax = axes[1][0])
sns.scatterplot(x="writing score", y="reading score", data= data, ax = axes[1][1])

sns.scatterplot(x="reading score", y="writing score", data= data, ax = axes[2][0])
sns.scatterplot(x="math score", y="writing score", data= data, ax = axes[2][1])
```
